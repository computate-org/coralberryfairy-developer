## Bash Kernel for Jupyter

For the rest of the course we will be running terminal commands directly from a Jupyter Notebook in VSCode. 
For this you will want to install the Bash Kernel for Jupyter. 
This is easy to install with python pip dependencies. 
Open a terminal in VSCode to the `coralberryfairy-developer` directory and run the commands below to install the Bash Kernel for Jupyter. 

## Configure VSCode extensions

There are some useful VSCode extensions that you can add to your VSCode workspace, that you can install to get started developing smart-aquaculture. 

Open a terminal in VSCode to the `coralberryfairy-developer` directory and run the commands below to configure the recommended extensions. 

```bash
install -d .vscode/
echo '{
  "recommendations": [
    "ms-toolsai.jupyter"
    , "redhat.java"
    , "redhat.vscode-yaml"
    , "vscjava.vscode-java-debug"
    , "vscjava.vscode-maven"
  ]
}
' | tee .vscode/extensions.json
echo DONE
```

To install these recommended extensions, 
- open the VSCode extensions menu on the left, 
- then click the Filter button, 
- select `Recommended` extensions, 
- and click the `Install` button on each of the recommended extensions. 

## Switch to the Bash Kernel

To run this Jupyter Notebook in VSCode, you will need to switch to the Bash Kernel you installed earlier. 
It's a button in the top right corner of the notebook that says `Select Kernel`. 


## Configure VSCode settings

To develop coralberryfairy, it's easiest to clone the `coralberryfairy-developer` project, 
then clone the `coralberryfairy` and `coralberryfairy-static` repos inside of `coralberryfairy-developer`. 
The `coralberryfairy-developer` project ignores the repos inside of it with the .gitignore file. 
So we will configure the VSCode settings for the workbench to find .gitignore files. 

```bash

install -d .vscode/
echo '{
    "python-envs.pythonProjects": [
        {
            "path": "",
            "envManager": "ms-python.python:venv",
            "packageManager": "ms-python.python:pip"
        }
    ],
    "search.useIgnoreFiles": false
}
' | tee .vscode/settings.json
echo DONE
```
