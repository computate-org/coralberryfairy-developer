## Switch to the Bash Kernel

To run this Jupyter Notebook in VSCode, you will need to switch to the Bash Kernel you installed earlier. 
It's a button in the top right corner of the notebook that says `Select Kernel`. 


## Install Java 21 Development Kit

Run the following commands to install the Java 21 Development Kit and make it the default. 
The last 2 commands require a sudo password to set Java 21 as the default `java` and `javac`. 


```bash
pkcon install -y maven
pkcon install -y java-21-openjdk-devel
echo DONE
```

These you need to run these separate commands in the terminal.

## Update JAVA_HOME in ~/.bashrc

If the default Java on your computate is Java 11 or earlier, 
you will want to update your JAVA_HOME environment variable to Java 17 in your `~/.bashrc` file. 

Then you will want to update your terminal closing and opening a new terminal, 
or run `source ~/.bashrc`. 

```bash

echo "export JAVA_HOME=/usr/lib/jvm/java-21-openjdk" | tee -a ~/.bashrc
echo DONE
```
